```python
import pandas as pd
file_path="C:\\Users\\Hadjer\\Desktop\\Interview\\Merge.csv"
df = pd.read_csv(file_path) 
df.head()
new_df = df[['Resolution', 'Datetime', 'Price']].copy()
new_df.head()
```

```python
new_df['Datetime'] = pd.to_datetime(new_df['Datetime'])
```

```python
data=new_df.set_index(['Datetime'])
data
```

```python
option1 = ['10MIN']
option2 = ['1H']
option3 = ['D']
df1= data.loc[data['Resolution'].isin(option1)]
df2= data.loc[data['Resolution'].isin(option2)]
df3= data.loc[data['Resolution'].isin(option3)]
```

```python
df1
```

```python
df2
```

```python
df1_price=pd.DataFrame(df1.Price.resample('2H').mean())
df1_price
```

```python
df1=df1_price.between_time('07:00', '17:00')
df1
```

```python
df2_price=pd.DataFrame(df2.Price.resample('2H').mean())
df2_price
```

```python
df2=df2_price.between_time('07:00', '17:00')
df2
```

Call resample('2H') to resample the given time-series by DAY. After that, ffill() is called to forward fill the values.

```python
df3_price = pd.DataFrame(df3.Price.resample('2H').ffill())
df3_price
```

```python
df3=df3_price.between_time('07:00:00', '17:00:00')
df3
```
