```python
import pandas as pd
file_path="C:\\Users\\Hadjer\\Desktop\\Interview\\Consumption.csv"
df = pd.read_csv(file_path) 
df
```

```python
mask = df[df['Date'].str.contains(r'2020')]
mask
```

```python
df.drop(df[df['Date'].apply(len) < 8].index, inplace = True)
df['dateformat'] = 1
df.loc[df.Date.str.contains('/'), 'dateformat'] = 2
mask = df[df['Date'].str.contains(r'2020')]
df
```

```python
df.loc[df.dateformat == 1, 'new_date'] = pd.to_datetime(df.loc[df.dateformat == 1, 'Date'], format = '%Y%m%d').dt.strftime('%d/%m/%Y')
df.loc[df.dateformat == 2, 'new_date'] = pd.to_datetime(df.loc[df.dateformat == 2, 'Date'], format = '%d/%m/%Y').dt.strftime('%d/%m/%Y')
df
```

```python
data = df[['new_date', 'Consumption']].copy()
data
```

```python
data['Year']= pd.to_datetime(data['new_date'], format = '%d/%m/%Y').dt.year
data['MonthDay']= pd.to_datetime(data['new_date'], format = '%d/%m/%Y').dt.strftime('%m-%d')
data
```

```python
df_pivot=data.pivot(index=['MonthDay'], columns='Year',values=['Consumption'])
df_pivot
```
